```python
import pandas as pd
import sentencepiece as spm

# Step 1: Read the CSV file
csv_file = '/content/Amharic News Dataset.csv'
df = pd.read_csv(csv_file)

# Step 2: Extract text data from the relevant column
text_data = df['article'].tolist()

# Step 3: Write the text data to a plain text file
with open('/content/data.text', 'w', encoding='utf-8') as f:
    for line in text_data:
        f.write(f"{line}\n")
```

```python
import sentencepiece as spm

# Train a SentencePiece model
spm.SentencePieceTrainer.train('--input=/content/data.text --model_prefix=m --vocab_size=26000')

# Load the trained model
sp = spm.SentencePieceProcessor(model_file='m.model')

# # Encode text into subwords
# text = "This is a test."
# tokens = sp.encode(text, out_type=str)
# print(tokens)  # ['This', 'is', 'a', 'test', '.']

# # Decode subwords back to text
# decoded_text = sp.decode(tokens)
# print(decoded_text)  # "This is a test."
```
